Plot dos resultados utilizando o modelo PTBR \/

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("top10_emotions_PTBR.csv")

# extrai emocao e score da coluna 'top1_emocao'
df['emocao'] = df['top1_emocao'].str.extract(r'([a-zA-Z]+)')[0]
df['score'] = df['top1_emocao'].str.extract(r'\(([\d.]+)\)')[0].astype(float)

top10 = df.sort_values('score', ascending=False).head(10)
top10 = top10[::-1]  # inverte a ordem p o grafico horizontal

plt.barh(top10['musica'], top10['score'])
plt.xlabel("Música")
plt.ylabel("Score")
plt.title("Top 10 músicas com maior score de emoção")

# adiciona rotulos de emocao nas barras
for i in range(len(top10)):
    plt.text(top10.iloc[i]['score'], i, top10.iloc[i]['emocao'])

#plt.tight_layout()
plt.show()
```
